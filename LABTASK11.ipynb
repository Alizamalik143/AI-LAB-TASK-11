Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB, GaussianNB, MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
```

Load Dataset

```python
data = pd.read_csv("diabetes.csv")
x = data.iloc[:, 0:-1]
y = data.iloc[:, -1]
```

Convert Object Columns to Integers

```python
cat_columns = x.select_dtypes(['object']).columns
for col in cat_columns:
    x[col] = pd.factorize(x[col])[0] + 1
```

Split Data into Train and Test

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, shuffle=True, random_state=42)
```

Bernoulli Naive Bayes

```python
BernNB = BernoulliNB()
BernNB.fit(X_train, Y_train)
Y_bpred = BernNB.predict(X_test)
b_accuracy = accuracy_score(Y_test, Y_bpred)
b_precision = precision_score(Y_test, Y_bpred, average='weighted', labels=np.unique(Y_bpred))
b_recall = recall_score(Y_test, Y_bpred, average='weighted', labels=np.unique(Y_bpred))
b_f1 = f1_score(Y_test, Y_bpred, average='weighted', labels=np.unique(Y_bpred))
```

Random Forest

```python
RF = RandomForestClassifier()
RF.fit(X_train, Y_train)
Y_rpred = RF.predict(X_test)
r_accuracy = accuracy_score(Y_test, Y_rpred)
r_precision = precision_score(Y_test, Y_rpred, average='weighted', labels=np.unique(Y_rpred))
r_recall = recall_score(Y_test, Y_rpred, average='weighted', labels=np.unique(Y_rpred))
r_f1 = f1_score(Y_test, Y_rpred, average='weighted', labels=np.unique(Y_rpred))
```

Gaussian Naive Bayes

```python
GausNB = GaussianNB()
GausNB.fit(X_train, Y_train)
Y_gpred = GausNB.predict(X_test)
g_accuracy = accuracy_score(Y_test, Y_gpred)
g_precision = precision_score(Y_test, Y_gpred, average='weighted', labels=np.unique(Y_gpred))
g_recall = recall_score(Y_test, Y_gpred, average='weighted', labels=np.unique(Y_gpred))
g_f1 = f1_score(Y_test, Y_gpred, average='weighted', labels=np.unique(Y_gpred))
```

Decision Tree

```python
Dtree = DecisionTreeClassifier()
Dtree.fit(X_train, Y_train)
Y_dpred = Dtree.predict(X_test)
d_accuracy = accuracy_score(Y_test, Y_dpred)
d_precision = precision_score(Y_test, Y_dpred, average='weighted', labels=np.unique(Y_dpred))
d_recall = recall_score(Y_test, Y_dpred, average='weighted', labels=np.unique(Y_dpred))
d_f1 = f1_score(Y_test, Y_dpred, average='weighted', labels=np.unique(Y_dpred))
```

Multinomial Naive Bayes

```python
MultiNB = MultinomialNB()
MultiNB.fit(X_train, Y_train)
Y_mpred = MultiNB.predict(X_test)
m_accuracy = accuracy_score(Y_test, Y_mpred)
m_precision = precision_score(Y_test, Y_mpred, average='weighted', labels=np.unique(Y_mpred))
m_recall = recall_score(Y_test, Y_mpred, average='weighted', labels=np.unique(Y_mpred))
m_f1 = f1_score(Y_test, Y_mpred, average='weighted', labels=np.unique(Y_mpred))
```

K-Nearest Neighbors

```python
KNN = KNeighborsClassifier()
KNN.fit(X_train, Y_train)
Y_kpred = KNN.predict(X_test)
k_accuracy = accuracy_score(Y_test, Y_kpred)
k_precision = precision_score(Y_test, Y_kpred, average='weighted', labels=np.unique(Y_kpred))
k_recall = recall_score(Y_test, Y_kpred, average='weighted', labels=np.unique(Y_kpred))
k_f1 = f1_score(Y_test, Y_kpred, average='weighted', labels=np.unique(Y_kpred))
```

Line Graph of All Scores

```python
plt.figure(figsize=(16,10))
x_labels = ['Bernoulli', 'Random Forest', 'Gaussian', 'Decision Tree', 'Multinomial', 'KNeighbors']
plt.plot(x_labels, [b_accuracy, r_accuracy, g_accuracy, d_accuracy, m_accuracy, k_accuracy], marker='o', label='Accuracy')
plt.plot(x_labels, [b_precision, r_precision, g_precision, d_precision, m_precision, k_precision], marker='o', label='Precision')
plt.plot(x_labels, [b_recall, r_recall, g_recall, d_recall, m_recall, k_recall], marker='o', label='Recall')
plt.plot(x_labels, [b_f1, r_f1, g_f1, d_f1, m_f1, k_f1], marker='o', label='F1')
plt.title("Scores of Applied Classifiers")
plt.legend()
plt.show()
```

Bar Graph of F1 Scores

```python
plt.figure(figsize=(16,10))
left = [1, 2, 3, 4, 5, 6]
height = [b_f1, r_f1, g_f1, d_f1, m_f1, k_f1]
tick_label = ['Bernoulli', 'Random Forest', 'Gaussian', 'Decision Tree', 'Multinomial', 'KNeighbors']
plt.bar(left, height, tick_label=tick_label, width=0.9, color=['#08737f','#00898a','#089f8f','#39b48e','#64c987','#92dc7e'])
plt.xlabel('Classifiers')
plt.ylabel('F1 Scores')
plt.title('F1 Scores of Applied Classifiers')
plt.show()
```

Train-Test Splitting

```python
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv("diabetes.csv")
X = data.iloc[:, 0:-1]
y = data.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=42)
```

Choose and Apply the Model

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

Testing/Predicting

```python
y_pred = model.predict(X_test)
```

Display Accuracy Score

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
